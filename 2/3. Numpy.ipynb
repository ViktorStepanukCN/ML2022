### Vectorization

Running linear algebra operation on precompiled and optimized function instead of using loops.

$wX + b$

Define 1 000 000 of weights, values and single bias.

```python
# generate random vector consisting of 1 000 000 components
w = np.random.rand(1000000)
x = np.random.rand(1000000)
b = 10
```

Define stopwatch function for measuring execution time.

```python
def stopwatch(func):
    # get time before execution
    start = time.time()
    # execute the function
    func()
    # get time after execution
    end = time.time()
    # print the results
    print(f'{1000 * (end - start)} ms')
```

Implementation using loops.

```python
def loop(w, x, b):
    # define variable for storing result
    z = 0.0
    # run dot product operation number by number
    for i in range(len(x)):
        z += w[i]*x[i]
    # add bias
    z += b
    return z
```

Implementation employing vectorization.

```python
def vectorization(w, x, b):
    # run dot product from numpy
    return np.dot(w,x) + b
```

Runtime execution of loop implementation.

```python
stopwatch(lambda: loop(w, x, b))
```

Runtime execution of vectorization implementation.

```python
stopwatch(lambda: vectorization(w, x, b))
```
